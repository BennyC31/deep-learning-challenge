## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
data_url = 'https://raw.githubusercontent.com/BennyC31/deep-learning-challenge/main/charity_data.csv'
# application_df = pd.read_csv("../Resources/charity_data.csv")
application_df = pd.read_csv(data_url)
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
app_df = application_df
app_df = app_df.drop(['EIN','NAME'], axis=1)
app_df.head()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
unq_df = app_df
unq_cols = unq_df.nunique(axis=0)
unq_cols
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
app_type_cnt = app_df.loc[:, 'APPLICATION_TYPE'].value_counts()
app_type_cnt
# app_type_cnt = app_df['APPLICATION_TYPE'].value_counts()
# app_type_cnt
```

```python
application_types_to_replace = list(app_type_cnt[app_type_cnt<200].index)
application_types_to_replace
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
class_cnt = app_df.loc[:, 'CLASSIFICATION'].value_counts()
class_cnt
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
class_cnt.loc[class_cnt > 1]
```

```python
classifications_to_replace = list(class_cnt[class_cnt<1000].index)
# classifications_to_replace
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for cls in classifications_to_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
app_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
app_df = pd.get_dummies(app_df, dtype=float)
app_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = app_df.IS_SUCCESSFUL.values
X = app_df.drop('IS_SUCCESSFUL', axis=1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state= 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled[0]
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
num_features = len( X_train_scaled[0])
hidden_layer1=70
hidden_layer2=30

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer1, input_dim=num_features, activation='relu'))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_layer2, activation='relu'))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
#  YOUR CODE GOES HERE
from google.colab import files
nn.save_weights('AlphabetSoupCharity_Optimization1.h5')
```
